```python
# Función para buscar y extraer datos de Spotify:

def get_spotify_data(sp, genre, year_range):
    results = sp.search(q=f'genre:{genre} year:{year_range}', type='album,track')
    Tracks = results['tracks']['items']
    Albums = results['albums']['items']

    data = []
    for item in Tracks + Albums:
        if item['type'] == 'track':
            year = item['album']['release_date'][:4]
        else:
            year = item['release_date'][:4]

        data.append({
            'Artist': item['artists'][0]['name'],
            'Genre': genre,
            'Type': item['type'],
            'Title': item['name'],
            'Year': year,
            'Id_spotify': item['id']
        })
    return data
```

```python
# Código para usar la API_KEY de Last_fm desde el archivo .env que nos permite ocultar las claves:

# Cargar las variables de entorno desde el archivo .env (Last_fm):
load_dotenv()

# Obtener la clave de API desde la variable de entorno (Last_fm):
API_KEY = os.getenv('API_KEY')

# Comprobar si la clave de API se ha cargado correctamente (Last_fm):
if API_KEY is None:
    print("Error: No se encontró la API_KEY en las variables de entorno.")
    exit()
```

```python
# Crear una lista con los artista obtenidos en Spotify para poder realizar la búsqueda en Last_fm:

artist_spotify = []
for index, row in df_spotify_sinDuplicados.iterrows():
    artist_spotify.append(row['Artist'])

# Mostrar la lista de artistas de Spotify:
print(artist_spotify)
```

```python
# Lista donde se guardarán los datos de cada artista:
all_data_lastfm = []

# Función para obtener datos de Last_fm:
def get_lastfm_data(artist_name):
    try:
        # Endpoint de Last_fm para obtener la información de cada artista:
        url_info = f'http://ws.audioscrobbler.com/2.0/?method=artist.getinfo&artist={artist_name}&api_key={API_KEY}&format=json'  
        # Realizar las solicitudes a la API de Last_fm:
        response_info = requests.get(url_info)
        # Verificar si las respuestas obtenidas son correctas:
        if response_info.status_code == 200 :
            data_info = response_info.json()

            # Comprobar si hay información del artista:
            if 'artist' in data_info:
                biography = data_info['artist']['bio']['summary']
                playcount = data_info['artist']['stats']['playcount']
                listeners = data_info['artist']['stats']['listeners']
            # Obtener artistas similares:
                url_similars = f'http://ws.audioscrobbler.com/2.0/?method=artist.getsimilar&artist={artist_name}&api_key={API_KEY}&format=json'
                response_similars = requests.get(url_similars)
                
                if response_similars.status_code == 200:
                    data_similars = response_similars.json()
                    if 'similarartists' in data_similars:
                        similars = [similar['name'] for similar in data_similars['similarartists']['artist']]
                    else:
                        similars = []
                else:
                    similars = []
                
                # Mostrar la información obtenida en Last_fm:
                print(f"Biografía de {artist_name}: {biography}")
                print(f"Reproducciones: {playcount}")
                print(f"Oyentes: {listeners}")
                print(f"Artistas similares: {', '.join(similars)}")
                
                # Agregar los datos a la lista:
                all_data_lastfm.append({
                    'Artist': artist_name,
                    'Biography': biography,
                    'Playcount,': playcount,
                    'Listeners,': listeners,
                    'Similars': ', '.join(similars)  # Almacenamos los artistas similares como una cadena de texto
                })
            else:
                print(f"No se encontró información sobre el artista {artist_name}. Respuesta: {data_info}")
        else:
            print(f"Error en la solicitud para {artist_name}. Código de estado: {response_info.status_code}")

    except Exception as e:
        print(f"Error al obtener datos para {artist_name}: {e}")

# Iterar sobre la lista de artistas de Spotify y obtener la información para cada uno de ellos en Last_fm:
for artists in artist_spotify:
    get_lastfm_data(artists)

# Crear un DataFrame a partir de la lista vacía (Last_fm):
df_last_fm = pd.DataFrame(all_data_lastfm)

# Mostrar el DataFrame de Last_fm:
print(df_last_fm)
```
