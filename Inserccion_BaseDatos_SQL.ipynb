INSERCCION DATOS SPOTIFY

```python
# Convertimos el archivo datos_spotify.csv en un DataFrame:
df_spotify = pd.read_csv('datos_spotify.csv')
df_spotify.columns = ['Artist', 'Genre', 'Type', 'Title','Year','Id_spotify']
df_spotify
```

```python
# Cargamos la tabla de Spotify en SQL:
engine = create_engine('mysql+mysqlconnector://root:AlumnasAdalab@localhost/MusicStream')

# Insertamos los datos desde el DataFrame en MySQL:
df_spotify.to_sql('Spotify', con=engine, if_exists='append', index=False)
print("Datos insertados exitosamente en la base de datos.")
```
